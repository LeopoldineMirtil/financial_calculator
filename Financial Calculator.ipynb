# Financial Calculator Project

```python
import math
import sympy
from sympy import symbols
from sympy.solvers import solve
import numpy as np

def fin_calc():
  print("Financial Calculator Menu: \n",
            "1. Monthly Mortgage Payment                4. Rate of Growth                       7. Solve Logarithmic Equations\n",
            "2. Retirement Investment Balance           5. Annuity with Monthly Growth          8. Convert TO Scientific Notation\n",
            "3. Time Required to Double Investment.     6. Annuity with Continuous Growth       9. Convert FROM Scientific Notation\n")
 

  while True:
    
    user_choice = int(input("Enter the number(1-9) of your choice from the menu:"))

    if user_choice == 1:

      def mortgage_pmt():

        p = float(input("Enter the mortgage principal: "))
        r = float(input("Enter the annual percentage rate as a decimal (ex. 12.50% = 0.125): "))
        t = float(input("Enter the number of years for the loan: "))

        num = (r/12)*(1+(r/12))**(12*t)
        den = (1+(r/12))**(12*t) - 1
        pmt = p*(num/den)

        print(f"Monthly mortgage payment = ${pmt:,.2f}")

      mortgage_pmt()
      break

    elif user_choice == 2:  

      def retirement():
        p = float(input("Starting balance amount: "))
        r = float(input("Annual percentage rate as a decimal (ex. 12.50% = 0.125): "))
        t = int(input("Number of investment years: "))
        monthly = float(input("Monthly contribution: "))

        annuity = p

        for a in range(12*t):
            annuity = annuity + monthly
            interest = annuity * r / 12
            annuity = annuity + interest

        print(f"Retirement investment ending balance = ${annuity:,.2f}")

      retirement()
      break

    elif user_choice == 3:  

      def double():

        r = float(input("Enter the annual rate as a decimal: \n"))
        t = math.log(2)/r

        #output
        print(f"Investment will double in = {t:,.2f} years")

      double()
      break

    elif user_choice == 4: 

      def growth_rate():
        p = int(input("Enter the starting value: "))
        e = int(input("Enter the ending balance: "))
        t = int(input("Enter the time in years: "))

        r = round((math.log(e/p)/t)*100, 2)

        print(f"Rate of growth = {r:.2f}%")
      growth_rate()
      break  

    elif user_choice == 5:  

      def anu_mo():

        p = float(input("Enter the mortgage principal: "))
        r = float(input("Enter the annual percentage rate as a decimal (ex. 12.50% = 0.125): "))
        t = float(input("Enter the number of years for the loan: "))

        #formula for 12 times annually
        mo = p*(1+r/12)**(12*t)

        print(f"Annuity with monthly growth = ${mo:,.2f}")

      anu_mo()
      break  

    elif user_choice == 6: 

      def an_cont():

        p = float(input("Enter the mortgage principal: "))
        r = float(input("Enter the annual percentage rate as a decimal (ex. 12.50% = 0.125): "))
        t = float(input("Enter the number of years for the loan: "))
        
        #calc continuous growth 
        con = p*math.e**(r*t)
        
        print(f"Annuity with continuous growth = ${con:,.2f}")

      an_cont()
      break  
    
    elif user_choice == 7:  

      def solve_logs():

          #set x to be solved for
          x = symbols('x')

          #input log function
          log_func = input("Enter logarithmic function: ")

          #solve function
          solution = solve(log_func,x)

          #output
          print(f'The solution for {log_func} is:  x = {solution}')

      solve_logs()
      break  

    elif user_choice == 8:  

      def to_scifi():  

        num = float(input('Enter the positive number to be converted to scientific notation: '))

        x = math.floor(math.log(num,10))
        n = round(num*10**(-x),2)

        print(f'{np.format_float_positional(np.float32(num), trim="-")} converts to: {n} x 10^{x}')

      to_scifi()
      break  

    elif user_choice == 9:  
    
      def from_scifi():
    
        #get number to be converted from scientific notation
        num = input("Enter positive number to be converted from scientific notation (ex. 2.56 x 10^6): ")

        #conversion FROM scientific
        ## split base and 10^exp
        base_pwr = num.split("x")
        base = float(base_pwr[0])

        #split exp from 10
        split_exp = str(base_pwr[1]).split("^")
        exp = int(split_exp[1])

        #converted number
        conv_num = base *10**exp

        #output
        print(f'{base} x 10^{exp} converts to: {np.format_float_positional(np.float32(conv_num), trim="-")}')

      from_scifi()
      break  

    else:
      print("ENTRY ERROR!! Number MUST be from 1 to 9")

fin_calc()
```
